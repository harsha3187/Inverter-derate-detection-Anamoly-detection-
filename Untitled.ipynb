```python
import pandas as pd
d_1 = pd.read_csv('s_data_all_rem_new.csv')
```

```python
d_1 = d_1.drop(columns=['Unnamed: 0'])
```

```python
d_1.head()
```

```python
d_1['Timestamp']=pd.to_datetime(d_1['Timestamp'])
```

```python
d_1.transpose().index
```

```python
import numpy as np
from sklearn.decomposition import PCA
abc = d_1[(d_1.Timestamp>=pd.to_datetime('2016-12-03 00:00:00' )) & (d_1.Timestamp<=pd.to_datetime('2016-12-04 00:00:00' ))][['INV1', 'INV2', 'INV3', 'INV4', 'INV5', 'INV6', 'INV7',
           'INV8', 'INV9', 'INV10', 'INV11', 'INV12', 'INV13', 'INV14', 'INV15',
           'INV16', 'INV17', 'INV18', 'INV19', 'INV20']].transpose().values.copy()
X = abc
pca = PCA(n_components= 2)
pca.fit(X)
X_pca=pca.transform(X)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
X_pca
```

```python
import matplotlib.pyplot as plt
Xax=X_pca[:,0]
Yax=X_pca[:,1]
labels=['INV1', 'INV2', 'INV3', 'INV4', 'INV5', 'INV6',
       'INV7', 'INV8', 'INV9', 'INV10', 'INV11', 'INV12', 'INV13', 'INV14',
       'INV15', 'INV16', 'INV17', 'INV18', 'INV19', 'INV20']

#cdict={0:'red',1:'green'}
# labl={0:'Malignant',1:'Benign'}
# marker={0:'*',1:'o'}
# alpha={0:.3, 1:.5}
fig,ax=plt.subplots(figsize=(7,5))
fig.patch.set_facecolor('white')
ax.scatter(Xax,Yax,s=40)
for i, txt in enumerate(labels):
    ax.annotate(txt, (Xax[i], Yax[i]))

# for l in np.unique(labels):
#  ix=np.where(labels==l)
 
# for loop ends
plt.xlabel("First Principal Component",fontsize=14)
plt.ylabel("Second Principal Component",fontsize=14)
plt.legend()
plt.show()
```
